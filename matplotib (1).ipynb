```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,6,11)
y=x**2

fig=plt.figure()

axes=fig.add_axes([0.01,0.01,0.3,0.3])

axes.plot(x,y,'b')
axes.set_xlabel('set x label')
axes.set_ylabel('set y label')
axes.set_title('set title')
```

```python
x=np.linspace(0,6,11)
y=x**2

fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.3,0.3])
axes2=fig.add_axes([0.15,0.25,0.1,0.1])
axes2.plot(x,y,'r-')
axes.plot(x,y,'b')
axes.set_xlabel('set x label')
axes.set_ylabel('set y label')
axes.set_title('set title')
```

```python
import pandas as pd
import numpy as np

df=pd.DataFrame(np.random.randint(1,100
,(4,4)),index=['riya','siya','kiya','jiya'],
columns=['python','Data_Science','DBMS','DSA'])

import matplotlib.pyplot as plt
fig1, axes1=plt.subplots(2,2,layout='constrained',sharex=True,sharey=True)

axes1[1,1].plot(df.index,df.DBMS,'gs:',label="DBMS")
axes1[1,1].legend()


axes1[1,0].plot(df.index,df.python,'gs:',label="Data_Science")
axes1[1,0].legend()

axes1[0,0].plot(df.index,df.Data_Science,'gs:',label="DSA")
axes1[0,0].legend()

axes1[0,1].plot(df.index,df.python,'gs:',label="python")
axes1[0,1].legend()
axes1[0,1].grid(True)
```
